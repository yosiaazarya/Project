```python
import matplotlib.pyplot as plt
import numpy as np

data = np.arange(10)
data2 = np.sin(data*np.pi)
data3 = 2 * np.sin(data*np.pi)
plt.figure()
plt.plot(data,data2, color='blue')
plt.plot(data,data3, color='orange')
plt.show()

fig = plt.figure()
ax1 = fig.add_subplot(2,3,1)
ax2 = fig.add_subplot(2,3,6
                     )
ax1.plot(data,data2,color='blue')
ax1.plot(data,data3,'g--')
plt.show()


plt.figure()
plt.subplot(221)
plt
```
